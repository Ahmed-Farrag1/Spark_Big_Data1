```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.feature import VectorAssembler
```

```python
sc=SparkSession.builder.appName("Decision tree regressor").getOrCreate()
```

```python
data = sc.read.option("header","true").option("inferSchema","true").csv(r"C:\Spark Course\Decision_Tree\realestate.csv")
```

```python
assembler =VectorAssembler().setInputCols(["HouseAge","DistanceToMRT","NumberConvenienceStores"]).setOutputCol("features")
```

```python
df=assembler.transform(data).select("PriceOfUnitArea","features")
print(df.take(1))
```

```python
train_test =df.randomSplit([.5,.5])
train=train_test[0]
test=train_test[1]
```

```python
dtr=DecisionTreeRegressor().setFeaturesCol("features").setLabelCol("PriceOfUnitArea")
model =dtr.fit(train)
prediction = model.transform(test).cache()
```

```python
df1= prediction.select("PriceOfUnitArea","prediction")
for i in df1.collect():
# 1- actual price  ---  2- prediction
    print((i[0],i[1]))
sc.stop()
```
